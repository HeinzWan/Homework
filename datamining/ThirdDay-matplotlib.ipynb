```python
fig= plt.figure(figsize=(12,3))
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)
ax1.scatter(iris['SepalLengthCm'],iris['SepalWidthCm']);
ax2.scatter(iris['PetalLengthCm'],iris['PetalWidthCm']);
```

```python
plt.scatter(iris['PetalLengthCm'],iris['PetalWidthCm']);
plt.title('petal size',size=20,color='y')
plt.xlabel('petal length',size=14,color='b')
plt.ylabel('petal width',size=12,color='r')
```

```python
plt.scatter(iris['PetalLengthCm'],iris['PetalWidthCm']);
```

```python
plt.scatter(iris['PetalLengthCm'],iris['PetalWidthCm']);
plt.xlim(0,8);
plt.ylim(0,3);
```

```python
#ticks
plt.scatter(iris['PetalLengthCm'],iris['PetalWidthCm']);
plt.xticks([1.5,4,6.5],fontsize=12,labels=['small','medium','large'])
plt.yticks([0.2,1.2,2.2],fontsize=12,labels=['small','medium','large'])
#listsmall medium large
```

```python
#legend
plt.scatter(iris.loc[iris.Species == 'Iris-setosa','PetalLengthCm'],\
            iris.loc[iris.Species == 'Iris-setosa','PetalWidthCm'],color='r',label='setosa')
plt.scatter(iris.loc[iris.Species == 'Iris-versicolor','PetalLengthCm'],\
            iris.loc[iris.Species == 'Iris-versicolor','PetalWidthCm'],color='b',label='versicolor')
plt.scatter(iris.loc[iris.Species == 'Iris-virginica','PetalLengthCm'],\
            iris.loc[iris.Species == 'Iris-virginica','PetalWidthCm'],color='g',label='virginica')
plt.xlim(0,8);
plt.ylim(0,3);
plt.legend();
```
