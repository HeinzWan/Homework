```python
def one_Embarked(data):
    embarked=pd.get_dummies(titanicDf['Embarked'])
    data = pd.concat([data,embarked],axis=1)
    return data

titanicDf=one_Embarked(titanicDf)
titanicDf
```

```python
def one_title(data):
    title=pd.get_dummies(titanicDf['Title'])
    data = pd.concat([data,title],axis=1)
    return data

titanicDf=one_title(titanicDf)
titanicDf
```
